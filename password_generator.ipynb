## Building The Password Generator Program

We have been learning about for loops, range, randomization, lists etc. Today, we combine all that we have learned so far to build a program that randomly generates a password for you by combining letters, numbers and symbols. We will be trying different methods until we achieve the best result.

For the first program, the user will be require dto input the number of characters they want in their password. The for loop allows you go over the items in a list one after the other, and do whatever you want with them. Randomization, just like the word, assigns random value(s) to you. To use it, you have to first import the random module.

```python
#import the random module
import random

#print a welcome message
print("Welcome to your Password Generator")

#create the list of options the user can choose from and assign it to a variable
letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x','y','z']
symbols = ['£', '@', '*', '$', '%', '&', '!', '#', '?', '+']
numbers = ['0', '2', '3', '4', '5', '6', '7', '8', '9']

#ask for the users input, it must be an integer so we typecast
num_letters = int(input("How many letters do you want in your password?\n"))
num_symbols = int(input("How many symbols do you want in your password?\n"))
num_numbers = int(input("How many numbers do you want in your password?\n"))

#create an empty string to house the password
your_password = ""

#iterate through the range of the users password length choice for each character
for character in range (0, num_letters): #the range starts from zero and ends in whatever number the user chose
    your_password += random.choice(letters) #the new password will be a random choice from the list of letters
for character in range (0, num_symbols):
    your_password += random.choice(symbols)
for character in range (0, num_numbers):
    your_password += random.choice(numbers)

print(your_password)
```

**Great!** However to make the password stronger and less hackable, it needs to be able to mix up the characters. That is, letters, numbers and symbols randomly mixed up. To do this, we can follow the same strategy but this time, we will be adding our password characters to a list. Afterwords, we will have to randomize the output!

We can use .append() to replace password += random.choice(letters). Logically, they function the same way. The += means password = password + random.choice(letters). Instead of repeating password twice, += is used. This is applied in every situation where you need to update the value of something after adding something else on top of it. append() does the same thing in list situations. Moving on...

**Mixing uppercase and lowercase**: I also want the letters to have an option of randomly mixing uppercase and lowercase in the password. My ideal password should look like this D&tyA%1hQ3*Ywez. To achieve the letter mixture, python has a module that allows you do that.

The "string" module. First, import string. Then, just as we did for random, call string to take on the action that you want.

A summary of what string module provides:

string.ascii_lowercase = 'abcdefghijklmnopqrstuvwxyz'

string.ascii_uppercase = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

string.ascii_letters = both lowercase and uppercase

string.digits = '0123456789'

```python
#import the necessary modules
import random
import string

print("Welcome to your Password Generator")
letters = list(string.ascii_letters)  # Contains both lowercase and uppercase
symbols = ['£', '@', '*', '$', '%', '&', '!', '#', '?', '+']
numbers = ['0', '2', '3', '4', '5', '6', '7', '8', '9']

num_letters = int(input("How many letters do you want in your password?\n"))
num_symbols = int(input("How many symbols do you want in your password?\n"))
num_numbers = int(input("How many numbers do you want in your password?\n"))

#so first, we create an empty list
password_list = [] #create an empty list to house the password
#iterate through the range of the users password length choice for each character
for character in range (0, num_letters): #the range starts from zero and ends in whatever number the user chose
    password_list.append(random.choice(letters)) #changed += to append, to add the random letters to our empty list
for character in range (0, num_symbols):
    password_list.append(random.choice(symbols))
for character in range (0, num_numbers):
    password_list.append(random.choice(numbers))

print(password_list) #this will still give us a similar answer as the code above because the output is still not random
#to randomize
random.shuffle(password_list)
print(password_list)
#our result is a list. to turn it back into a string, we will use a for loop

password = "" #create an empty string to house the new password
for character in password_list:
    password += character
print(password)
```

Did you see the difference using a list and randomizing our output made? Lets take it a step further. We do not want to bother our users by asking them for input. To automate the password generation process, we will assign a range to our characters. We decide the length of the password and write the program to randomly generate it.

```python
# Alternatively, we can skip the whole process of asking the user for input, decide what the length of a strong password is, and run with that
import random
import string

print("Welcome to your Password Generator")

letters = list(string.ascii_letters) #gets all letters regardless of case
symbols = ['£', '@', '*', '$', '%', '&', '!', '#', '?', '+']
numbers = list(string.digits) #gets digits 0,9

#our chosen length of characters
num_letters = 8
num_symbols = 3
num_numbers = 3

password_list = []  # create your empty list

for character in range(num_letters):
    password_list.append(random.choice(letters))
    
for character in range(num_symbols):
    password_list.append(random.choice(symbols))
    
for character in range(num_numbers):
    password_list.append(random.choice(numbers))

random.shuffle(password_list)  # Shuffle the password order!

your_password = "".join(password_list)  # did you see our new function 'join'? takes away the for loop we used earlier to convert list to string

print(f"Your generated password is: {your_password}")
```
